```python
#создание различных массивов из обычного списка и из специальных функций
import numpy as np

array = np.array([1,2,3,4,5]) #создаем одномерный массив
print('array')
print(array)

array1 = np.ones((2,1,3)) #создаем массив единиц размером (2,1,3)
print('array1')
print(array1)

array2 = np.eye(5) #создаем единичную матрицу размером (5,5)
print('array2')
print(array2)

array3 = np.zeros((3,3)) #создаем матрицу с нулями  размером (3,3)
print('array3')
print(array3)

array4 = np.arange(10,50,5) #создаем массив от 10 до 50 с шагом 5
print('array4')
print(array4)

array5 = np.empty((4,2,1)) #создаем небольшую трехмерную матрицу, заполненную мусором
print('array5')
print(array5)

array6 = np.arange(1,1000,0.0001) #создаем массив с большим количеством элементов
print('array6')
print(array6)
```

```python
#печатаем полученный массивы
print(array.shape)
print(type(array))
print(array1.shape)
print(type(array1))
```

```python
# задаем параметры для вывода большого массива
np.set_printoptions(threshold=100, precision=10)
print(array6)
np.set_printoptions(threshold=1000, precision=8) #затем возвращаем стандартные настройки
```

```python
#Базовые операции с массивами
a = np.array([4,5,6,7,8]) #первый массив
print(a)
b = np.eye(2) #второй массив
print(b)

#сложим их и получим ошибку, так как размеры не совпадают
#print(a+b)

c = np.arange(5)
print(c)

print(a+c) #сложение
```

```python
print(a-c) #вычитание
```

```python
print(a*c) #поэлементное умножение
```

```python
print(a/c) #поэлементное деление
#вернет ошибку с делением на 0, но программа вернула массив результата с элементом inf
```

```python
mat = np.array([[1,2,3,4],
               [11,12,13,14],
               [21,22,23,24],
               [31,32,33,34]])
mat
```
